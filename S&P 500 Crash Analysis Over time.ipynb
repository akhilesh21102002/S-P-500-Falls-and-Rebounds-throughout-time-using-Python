```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('SP500_Top10Falls_and_Rebounds.xlsx')
```

```python
df["Label"] = df["Date"].dt.strftime("%Y-%m-%d")
```

```python
df
```

```python
sns.set(style = "whitegrid")
plt.figure(figsize = (14,8))
```

```python
bar_width = 0.4
index = range(len(df))
```

```python
plt.bar(index, df["One_Day_Return_%"],width = bar_width, color = "crimson", label = "1-Day Return (%)")

plt.bar([i + bar_width for i in index],df["One_Year_Return_%"], width = bar_width, color = "seagreen", label = "1-Year Return (%)")


plt.xlabel("Date", fontsize = 12)
plt.ylabel("Return (%)", fontsize=12)
plt.title("S&P 500: Worst 1-Day Falls vs 1-Year Returns", fontsize=16)
plt.xticks([i + bar_width / 2 for i in index], df["Label"], rotation=45, ha="right")
plt.axhline(0, color='gray', linewidth=0.7)
plt.legend()
plt.tight_layout()
plt.show()


plt.savefig("SP500_Chart.png", dpi=300, bbox_inches="tight")
```
